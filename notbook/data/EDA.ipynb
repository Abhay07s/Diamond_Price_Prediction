```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,6))
x=0
for i in numerical_columns:
    sns.histplot(data=df,x=i,kde=True)
    print('\n')
    plt.show()
```
